# Diagnostic Metrics

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.svm import SVC
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report,f1_score
```

```python
aca2=pd.read_csv("aca2_dataset_training.csv")
aca2_valid=pd.read_csv("aca2_dataset_validation.csv")
```

```python
label=preprocessing.LabelEncoder()
label.fit(np.unique(list(aca2['Activity'].values)))
aca2['Activity']=label.transform(list(aca2['Activity'].values))
label.fit(np.unique(list(aca2['SCHOOL'].values)))
aca2['SCHOOL']=label.transform(list(aca2['SCHOOL'].values))
label.fit(np.unique(list(aca2['Class'].values)))
aca2['Class']=label.transform(list(aca2['Class'].values))
```

```python
label.fit(np.unique(list(aca2_valid['Activity'].values)))
aca2_valid['Activity']=label.transform(list(aca2_valid['Activity'].values))
label.fit(np.unique(list(aca2_valid['SCHOOL'].values)))
aca2_valid['SCHOOL']=label.transform(list(aca2_valid['SCHOOL'].values))
label.fit(np.unique(list(aca2_valid['Class'].values)))
aca2_valid['Class']=label.transform(list(aca2_valid['Class'].values))
```

```python
aca2=aca2.drop('UNIQUEID', axis = 1)
aca2=aca2.drop('CODER', axis = 1)
aca2=aca2.drop('TRANSITIONS', axis = 1)
aca2=aca2.drop('FORMATchanges', axis = 1)
aca2=aca2.drop('STUDENTID', axis = 1)
aca2
```

```python
aca2_valid=aca2_valid.drop('UNIQUEID', axis = 1)
aca2_valid=aca2_valid.drop('CODER', axis = 1)
aca2_valid=aca2_valid.drop('TRANSITIONS', axis = 1)
aca2_valid=aca2_valid.drop('FORMATchanges', axis = 1)
aca2_valid=aca2_valid.drop('STUDENTID', axis = 1)
aca2_valid
```

## decision tree

```python
from sklearn.tree import DecisionTreeClassifier
aca2_tree = aca2.drop('ONTASK', axis = 1)
ONTASK = aca2['ONTASK']
aca2TreeModel = DecisionTreeClassifier()
aca2TreeModel.fit(aca2_tree, ONTASK)
```

```python
ontask_pred = aca2TreeModel.predict(aca2_tree)
performance = [item in ontask_pred for item in ONTASK]
print('The accuracy is', sum(performance)/len(performance)*100, '%')
```

```python
aca2_valid_tree = aca2_valid.drop('ONTASK', axis = 1)
ONTASK_valid = aca2_valid['ONTASK']
```

```python
ontask_valid_pred = aca2TreeModel.predict(aca2_valid_tree)
```

```python
from sklearn.metrics import accuracy_score
ac_tree = accuracy_score(ONTASK_valid, ontask_valid_pred)
print("The accuracy for decision tree", ac_tree*100, "%")
```
